## Draw Stock Chart of Microsoft

We use yfinance and matplot lib to draw the stock chart of Microsoft.

```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import date

# Define the ticker symbol for Microsoft
ticker_symbol = "MSFT"

# Fetch historical stock data from Yahoo Finance
msft_stock = yf.download(ticker_symbol, start="2013-01-01", end=date.today().isoformat(), auto_adjust=False)

# Plot the stock chart
plt.figure(figsize=(12, 6))
plt.plot(msft_stock.index, msft_stock["Close"], label="Microsoft (MSFT)", color="blue")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.title("Microsoft Stock Chart")
plt.legend()
plt.grid(True)
plt.show()
```

## Get times-series data for USA Gross Domestic Product (FRED/GDP)

We use Nasdaq-Data-Link and matplotlib to visualize the GDP of the USA.

The Nasdaq Data Link Python module is free but you must have a Nasdaq Data Link API key in order to download data.

Please use the following code snippet to pass your API key: `nasdaqdatalink.read_key(filename="/path/.your-nasdaq-data-link-apikey")`.

- [Nasdaq-Data-Link API Documentation](https://docs.data.nasdaq.com/)
- [Authentication](https://docs.data.nasdaq.com/docs/python-installation#authentication)

```python
import nasdaqdatalink
from datetime import date

data = nasdaqdatalink.get("FRED/GDP", start_date="1945-01-01", end_date=date.today().isoformat())

# plot
import matplotlib.pyplot as plt
data.plot(color='orange')
plt.ylabel('Billions of Dollars')
plt.xlabel('Year')
plt.title('USA - Gross Domestic Product (FRED/GDP)')
plt.show()
```
